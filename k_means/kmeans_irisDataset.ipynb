```python
from sklearn.datasets import load_iris
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
import pandas as pd
```

```python
# X = pd.DataFrame(iris.data)
# X.columns = iris.feature_names
X = iris.data
y = iris.target
```

```python
sns.scatterplot(x=X[:,0],y=X[:,2],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
```

```python
wcss=[]
for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(1,11) , y = wcss, marker="o")
```

```python
kmeans = KMeans(n_clusters = 3)
labels = kmeans.fit_predict(x_scaled)

sns.scatterplot(x=X[:,0],y=X[:,2],c=labels)
```

```python
# sns.scatterplot(x=X[:,0],y=X[:,2],c=y)
```

```python
from sklearn.decomposition import PCA

pca = PCA(2)
pca_x = pca.fit_transform(x_scaled)
```

```python
kmeans = KMeans(n_clusters = 3)
labels = kmeans.fit_predict(pca_x)

sns.scatterplot(x=pca_x[:,0],y=pca_x[:,1],c=labels)
sns.scatterplot(x=kmeans.cluster_centers_[:,0],y = kmeans.cluster_centers_[:,1],marker="x",c="red",s=200)
```
